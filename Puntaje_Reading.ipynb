```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#Mejor puntaje en "reading" (mejor 10% de los puntajes)

```python
df=pd.read_csv('/content/cleanstudentscomplete.csv')
df.drop(['Unnamed: 0'],inplace=True, axis='columns')
df=df.sort_values('reading_score',ascending=False).head(int(df['Student ID'].count()*.1)).reset_index()
sexo=pd.get_dummies(df[['gender']])
df=pd.concat([df,sexo],axis=1)
df=df.groupby('school_name')[('gender_F','gender_M')].sum()
df.plot(kind='bar', color=['violet','orange'],figsize=(15,9),xlabel='Escuela',ylabel='Numero de Alumnos')
if(sum(df['gender_F'])<sum(df['gender_M'])):
  print('Los hombres tienen mejor puntaje en reading')
else:
  print('Las mujeres tienen mejor puntaje en reading')
```

#Peor puntaje en "reading" (peor 10% de los puntajes)

```python
df1=pd.read_csv('/content/cleanstudentscomplete.csv')
df1.drop(['Unnamed: 0'],inplace=True, axis='columns')
df1=df1.sort_values('reading_score').head(int(df1['Student ID'].count()*.1)).reset_index()
sexo=pd.get_dummies(df1[['gender']])
df1=pd.concat([df1,sexo],axis=1)
df1=df1.groupby('school_name')[('gender_F','gender_M')].sum()
df1.plot(kind='bar', color=['violet','orange'],figsize=(15,9),xlabel='Escuela',ylabel='Numero de Alumnos')
if(sum(df1['gender_F'])<sum(df1['gender_M'])):
  print('Los hombres tienen peor puntaje en reading')
else:
  print('Las mujeres tienen peor puntaje en reading')
```
