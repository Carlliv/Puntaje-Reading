```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/cleanstudentscomplete.csv',usecols=['Student ID','student_name','gender','grade','school_name','reading_score','math_score'],index_col=['Student ID'])
a=[]
m=[]
for x in df['reading_score']:
  if x>=70:
   a.append('Aprobado')
  else:
    a.append('Reprobado')
for x in df['math_score']:
  if x>=70:
   m.append('Aprobado')
  else:
    m.append('Reprobado')
df.insert(6,'cat_read',m)
df.insert(7,'cat_math',m)
df
```

```python
sexo=pd.get_dummies(df[['gender']])
df=pd.concat([df,sexo],axis=1)
df=df.groupby(['cat_read',	'cat_math'])[('gender_F','gender_M')].sum()
df.plot(kind='bar', color=['yellow','orange'],figsize=(15,9),xlabel='Escuela',ylabel='Numero de Alumnos')
```
